# Get the data

## Download the Data


```python
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request

def load_housing_data():
    tarball_path = Path("datasets/housing.tgz")
    if not tarball_path.is_file():
        Path("datasets").mkdir(parents=True, exist_ok=True)
        url = "https://github.com/ageron/data/raw/main/housing.tgz"
        urllib.request.urlretrieve(url, tarball_path)
        with tarfile.open(tarball_path) as housing_tarball:
            housing_tarball.extractall(path="datasets")
    return pd.read_csv(Path("datasets/housing/housing.csv"))

housing = load_housing_data()
```

**Understanding this Data**
* Scaling and Capping: The median income values in the dataset have undergone a scaling process. The data has been transformed and adjusted to fit within a specific range. The upper limit has been set at 15 (specifically, 15.0001) for higher median incomes, and the lower limit is 0.5 (specifically, 0.4999) for lower median incomes. This means that any original values above 15 have been replaced with 15, and any original values below 0.5 have been replaced with 0.5.

* Representation: The scaled values of the median income attribute in the dataset are not expressed in US dollars (USD). Instead, they represent approximate tens of thousands of dollars. For example, a value of 3 in the dataset corresponds to approximately $30,000.

* Preprocessing in Machine Learning: Working with preprocessed attributes, such as the scaled and capped median income, is common in machine learning. It allows for consistency and comparability across different features in the dataset.
